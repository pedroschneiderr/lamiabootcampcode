# Importando os recursos

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import cross_val_score
from tensorflow.keras import backend as k
from tensorflow.keras import utils
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, InputLayer
from scikeras.wrappers import KerasClassifier
```

```python
df = pd.read_csv('Deep Learning com Python de A a Z/Parte 1 - Redes Neurais Artificiais/classificação mais classes/iris.csv')
df.head()
```

```python
X = df.iloc[:, 0:4].values #todas as linhas da colunas 0 a 4, convetendo a Numpy array, pois é o formato de entrada esperado pelas redes Keras
y = df.iloc[:, 4].values #todas as linhas da coluna 5
```

```python
X
```

```python
y
```

# Configurando os dados

Como as redes neurais Keras recebem e retornam valores numéricos, essas strings vão ter que ser codificadas:

```python
le = LabelEncoder()
y = le.fit_transform(y)
y
```

Como são 3 classes, a camada de saída vai ter 3 neurônios. Ou seja, cada classe vai ser representada por uma configuração em que 1 neurônio fica bem ativo e todos os outros ficam quase inativos.<br>
Essa atividade é determinada por uma probabilidade, com 1 neurônio (que representa uma classe) sendo muito provável e todos os outros (que representam outras classes) sendo muito improváveis.<br>
Essa configuração é implementada utilizando a função Softmax.

Para representar cada classe em função de 3 neurônios:

```python
y = utils.to_categorical(y)
y
```

Cada linha é uma amostra e cada coluna é referente à uma classe.<br>
Se tem 0, então a amostra não pertence àquela classe.<br>
Se tem 1, então pertence.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25) #25% para teste
```

# Estrutura e treinamento da Rede Neural

```python
def create_network():
    k.clear_session()
    network = Sequential()
    network.add(InputLayer(shape=(4,)))
    network.add(Dense(units = 4, activation = 'relu')) # units = (len(input)+len(output))/2
    network.add(Dense(units = 4, activation = 'relu'))
    network.add(Dense(units = 3, activation = 'softmax')) # camada de saída

    network.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['categorical_accuracy']) # mesmo tipo de função de perda e métrica, mas pra multiclasse
    return network
```

```python
network = create_network()
```

```python
network.fit(X_train, y_train, batch_size = 10, epochs=1000)
```

# Analisando a precisão da Rede

```python
predictions = network.predict(X_test)
```

```python
predictions
```

Cada valor é a probabilidade da amostra representada por aquela lista ser da respectiva classe.<br>
Para obtermos a classificação mais provável para cada amostra:

```python
predictions = predictions > 0.5
predictions
```

```python
y_test
```

Para convertermos cada amostra numa codificação mais simples da classe mais provável:

```python
y_test2 = [np.argmax(i) for i in y_test]
y_test2
```

```python
predictions = [np.argmax(i) for i in predictions]
predictions
```

Agora, para termos a precisão exata das previsões:

```python
accuracy = accuracy_score(y_test2, predictions)
accuracy
```

```python
conf_matrix = confusion_matrix(y_test2, predictions)
print('(0)setosa  (1)versicolor  (2)virginica')
conf_matrix
```

Apenas 1 erro ocorreu, que foi a classificação de uma versicolor em viginica

Alternativamente, pode-se utilizar o método de evaluação padrão da rede neural:

```python
network.evaluate(X_test, y_test)
```

# Aplicando Cross Validation

```python
network2 = KerasClassifier(model = create_network, epochs = 250, batch_size=10)
```

```python
results = cross_val_score(estimator = network2, X=X, y=y, cv = 10, scoring='accuracy')
```
