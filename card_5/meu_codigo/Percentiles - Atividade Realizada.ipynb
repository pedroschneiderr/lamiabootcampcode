# Percentiles

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

vals = np.random.normal(0, 0.5, 10000)

plt.hist(vals, 50)
plt.show()
```

```python
np.percentile(vals, 50)
```

```python
np.percentile(vals, 90)
```

```python
np.percentile(vals, 20)
```

Como a distribuição é uniforme com valores variando de 0 a 99, o X° percentil tende a ser próximo de X, diferentemente da distribuição normal com valores variando fora da escala percentual.

```python
alpha = 3
vals3 = np.random.pareto(alpha, 1000)#gera um array c/ 1000 valores distribuidos exponencialmente, c/ o expoente sendo alpha
plt.hist(vals3, 200) #gera um histograma de 50 barras com os valores de vals3
plt.show() #plota o histograma
```

```python
np.percentile(vals3, 50)
```

```python
np.mean(vals3)
```

Agora o 50° percentil está distante da média, pois se trata de uma distribuição exponencial.

```python
(np.percentile(vals3, 75) - np.percentile(vals3, 25))/(vals3.max() - vals3.min())
```

Ou seja, a Amplitude Interquartil é apenas ~4% da amplitude total dos dados, já que há uma concentração muito grande da distribuição em uma região do intervalo.

```python
(np.percentile(vals, 75) - np.percentile(vals, 25))/(vals.max() - vals.min())
```

Neste caso, a Amplitude Interquartil já é ~18% da total, já que os valores de vals estão distribuídos normalmente.
