# Média e Mediana

```python
import numpy as np
incomes = np.random.normal(27000, 15000, 10000)
```

Cria-se um array de 10000 floats que se distribuem normalmente em torno de 27000, com desvio padrão de 15000.

```python
np.mean(incomes) #método mean retorna a média dos valores no array passado como argumento
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.hist(incomes, 50) #cria um histograma de 50 barras com os dados de incomes 
plt.show() #plota o histograma gerado por último
```

"%matplotlib inline" é uma magic function que garante que o gráfico gerado seja mostrado logo abaixo da célula em que a função de plot foi chamada.

pyplot é um módulo da biblioteca matplotlib que permite a plotagem de dados de forma semelhante ao MATLAB.

```python
np.median(incomes) #retorna a mediana do conjunto de dados do array incomes
```

Neste caso, a mediana ficou próxima à média pela razão de que se trata de uma distribuição normal.

```python
incomes = np.append(incomes, [1000000000]) #adiciona-se um elemento de valor 1000000000 ao fim de incomes
```

```python
np.median(incomes)
```

A mediana não mudou muito, já que só foi adicionado um elemento ao array, o que não deslocou muito o centro do conjunto ordenado.

```python
np.mean(incomes)
```

Já a média aumentou bastante, já que o valor adicionado é muito maior do que os outros (um outlier) e deformou a média para longe do valor típico do array (que deve ser em torno de 27000).

## Normal (Gaussiana)

```python
from scipy.stats import norm #importa o módulo norm do módulo stats da biblioteca scipy
x = np.arange(-3, 3, 0.001) #gera um array com floats de -3 a 3 com intervalo 0.001
y = np.arange(-12, -6, 0.001)
plt.plot(x, norm.pdf(x))
```

O método plot do módulo Pyplot plotou um gráfico de função de densidade de probabilidade dos valores de x.

O primeiro argumento (array x) determinou os valores do eixos das abscissas. O segundo (norm.pdf(x)) determinou os valores do eixo das ordenadas com o retorno do método pdf do módulo norm que por si só recebeu x como argumento e retornou as probabilidades para cada valor no array de modo que formasse uma distribuição normal.

### Gerando floats aleatórios a partir de uma distribuição normal

Fazendo o caminho contrário do método anterior (gerando os valores a partir do método de plotagem).

```python
mu = 5.0 #a distribuição terá média 5.0
sigma = 2.0 #a distribuição terá desvio padrão 2.0
values = np.random.normal(mu, sigma, 10000) #gera os valores de uma distribuição normal com média mu e desvio padrão sigma
plt.hist(values, 50) #gera um histograma de 50 barras com os valores do array values
plt.show() #plota o histograma gerado
```

# Função de Massa de Probabilidade

## Binomial

```python
from scipy.stats import binom #importa o módulo binom
n, p = 10, 0.5 #o número de experimentos será 10 e a probabilidade de sucesso será 0.5
x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x, n, p))
```

plt.plot plotou um gráfico com abscissas sendo os valores do array x e as ordenadas sendo as probabilidades de cada um baseado numa distribuição normal com n (10) experimentos e probabilidade de sucesso p (0.5), geradas pela função pmf do módulo binom que é uma Função de Massa de Probabilidade.

 # Correlação e Covariância

## Implementação de Covariância

```python
#calcula e retorna o vetor de diferenças da média dos valores de x
def mean_dev(x):
    xmean = np.mean(x)
    return [xi - xmean for xi in x]

#calcula e retorna a covariância entre os valores de x e y
def covariance (x, y):
    n = len(x)
    return np.dot(mean_dev(x),mean_dev(y))/(n-1)

pageSpeeds = np.random.normal(3.0, 1.0, 1000) 
purchaseAmount = np.random.normal(50, 10.0, 1000)

sns.scatterplot(x=pageSpeeds, y=purchaseAmount)
```

```python
covariance(pageSpeeds, purchaseAmount)
```

Covariância muito próxima de 0, ou seja, não há relação entre as variáveis.

```python
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds #estabelece uma relação entre os conjuntos de dados
sns.scatterplot(x=pageSpeeds, y=purchaseAmount)
```

```python
covariance(pageSpeeds, purchaseAmount)
```

Agora há um valor longe de 0, pois foi estabelecida uma relação entre os valores das 2 variáveis.

## Implementação de correlação

```python
#retorna a correlação entre os dados dos arrays x e y
def correlation(x, y):
    sdx = np.std(x)
    sdy = np.std(y)
    return covariance(x, y) / sdx / sdy 
```

```python
correlation(pageSpeeds, purchaseAmount)
```

O resultado indica uma correlação negativa considerável entre as 2 variáveis. Ou seja, o aumento no tempo de carregamento das páginas de um site tende a ser acompanhado pela diminuição da quantidade de compras.

# Probabilidade Condicional

```python
np.random.seed(0) #faz com que os números aleatórios gerados a partir daqui sejam sempre os mesmos em todas as execuções futuras

totals = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0} #dicionario que armazenará o número de pessoas por década de idade
purchases = {20:0, 30:0, 40:0, 50:0, 60:0, 70:0} #dicionário que armazenará o número de compras feitas por pessoas de cada década
totalPurchases = 0 #variável que armazenará o número de compras feitas no total, por qualquer pessoa

for _ in range(100000): #loop que cria uma pessoa e possivelmente uma compra, 100000 vezes
    ageDecade = np.random.choice([20, 30, 40, 50, 60, 70]) #escolhe uma década aleatóriamente
    purchaseProb = float(ageDecade)/100.0 #define a probabilidade de compra, quanto mais velho mais chance de comprar
    totals[ageDecade] += 1
    if np.random.random() <= purchaseProb: #random.random() gera um n° aleatório entre 0 e 1
        #ou seja, tem (purchaseProb*100)% de chance da execução entrar nesse bloco
        totalPurchases += 1
        purchases[ageDecade] += 1
```

### Computando P(E|F), com E sendo "comprar" e F sendo "pessoa nos 30".

Isso significa dizer a probabilidade de uma pessoa comprar algo sabendo que ela está na casa dos 30 anos.

Ou seja, é P(E,F)/P(F)

```python
PEF = float(purchases[30])/float(totals[30]) #é necessário garantir que o resultado seja um float pois probabilidade é sempre <= 1
print('P(E|F) = ' + str(PEF))
```

P(F) é a probabilidade de se estar na casa dos 30:

```python
PF = float(totals[30])/100000.0 #n° de pessoas na casa dos 30 dividido pelo n° total de pessoas
print('P(F) = ', str(PF))
```

P(E) é a probabilidade de alguém comprar algo independentemente da idade:

```python
PE = float(totalPurchases)/100000.0 #número de compras total dividido pelo número de pessoas
print('P(E) = ', str(PE))
```

P(E,F) é a probabilidade de alguém ter 30 anos e ter comprado algo:

```python
PEandF = float(purchases[30])/100000.0 #numero de pessoas com 30 anos que compraram algo dividido pelo total de pessoas
print('P(E, F) = ', str(PEandF))
```

Como E e F são eventos dependentes, P(E, F) é diferentes de P(E)*P(F):

```python
PE*PF
```
